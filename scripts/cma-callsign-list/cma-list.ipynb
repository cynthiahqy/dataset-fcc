# Scrape list of Cellular Market Areas (CMA)

saved dropdown list element from: https://wireless2.fcc.gov/UlsApp/UlsSearch/searchMarket.jsp
on 3/4/2019 into cma-element.html

```python
# https://www.crummy.com/software/BeautifulSoup/bs4/doc/#quick-start
from bs4 import BeautifulSoup
with open("cma-element.html") as cma:
    soup = BeautifulSoup(cma, "lxml")
```

```python
print(soup.prettify())
```

```python
soup.get_text("|", strip=True)
```

```python
lst = soup.get_text("|", strip=True).split("|")
```

```python
lst
```

```python
with open("cma-list.csv", "w") as f:
    for line in lst:
        f.write(line + "\n")
```

```python
with open("cma-list.csv", "r") as f:
    for line in f.readlines():
        print(line)
```
